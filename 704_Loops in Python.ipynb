# Loops in python

In generl, statemenbts are exeuted sequentially: The first statement in a function is executed first
Followed by the second, and so on. There may be a solution when you need to do execute a block of code several number of times.
programming languages provide various control structures that allow more complicated execution paths
a loop statement allow us to execute a statement or group of statements multiple times. The fllowing diagram illustrate a loop staements.

```python
student_name = str(input())
marks = {'Savri': 701, 'Jayesh': 705, 'Chandan': 708}
for student in marks:
    if student == student_name:
        print(marks[student])
        break
else:
     print('No entry with that name found')                    
```

```python
# While Loops
n=int(input("Enter n:"))

#n=10

#initialize sum and counter
sum=0
i=1
while i <= n:
    sum = sum + i
    i=i+1  #update counter
# print the sum
print("The sum is", sum)
```
